```python
from sklearn import tree
```

```python
classifier = tree.DecisionTreeClassifier()
```

```python
X = [[0, 0], [1, 1]]
y = [0, 1]
```

```python
classifier = classifier.fit(X, y)
```

```python
classifier.predict([[2., 2.]])
```

```python
classifier.predict_proba([[2., 2.]])
```

```python
tree.plot_tree(classifier)
```
